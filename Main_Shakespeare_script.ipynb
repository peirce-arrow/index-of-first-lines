```python
r = requests.get('https://www.folger.edu/explore/shakespeares-works/shakespeares-sonnets/read/')
```

```python
html_code = r.content
soup = BeautifulSoup(html_code, 'lxml')
```

```python
html = soup.decode()
```

```python
type(html)
```

```python
#lines = re.findall(r'Son-[0-9]{3}-[0-9]{2}.+<', html)
lines = re.findall(r'Son-[0-9]{3}-01.+<br', html)
for i in range(len(lines)):
    lines[i] = lines[i].replace('\xa0', u' ').replace('<br','')
    lines[i] = re.sub(r'Son-[0-9]{3}-01"></span>', f'{i} ', lines[i])
    lines[i] = re.sub(r'<span.+fdt-emendation-halfsquare.+>⌜</span>','', lines[i])
    lines[i] = re.sub(r'<span.+fdt-emendation-halfsquare.+>⌝</span>','', lines[i])


    #lines[i] = lines[i].replace('<span class="fdt-emendation-halfsquare" alt="" title="">⌜</span>','')
    #lines[i] = lines[i].replace('<span class="fdt-emendation-halfsquare" alt="" title="">⌝</span>','')
    #'<span alt="" class="fdt-emendation-halfsquare" title="">⌜</span>'
    #'<span alt="" class="fdt-emendation-halfsquare" title="">⌝</span>''
print(lines)
```

```python
split_words = []
end_words = []
for i in lines:
    split_words.append(i.split(' '))
for i in range(len(split_words)):
    end_words.append([])
    if i < len(split_words):
        end_words[i].append(split_words[i][0])
        end_words[i].append(split_words[i][-1])
print(end_words)
```

```python
for i in end_words:
    i[1] = i[1].translate(str.maketrans('', '', string.punctuation))

print(end_words)
```

```python
for i in end_words:
    if i[1].endswith("’d"):
        i[1] = i[1].replace("’","e")
        print(i[1])
    elif i[1].endswith("’er"):
        i[1] = i[1].replace("’","v")
        print(i[1])
    elif i[1].endswith("’st"):
        i[1] = i[1].replace("’","e")
        print(i[1])
    elif i[1] == 'recured':
        i[1] = i[1].replace("ured","urred")
        print(i[1])
    elif i[1] == '’greeing':
        i[1] = i[1].replace("’","a")
        print(i[1])
    elif i[1] == 'blest':
        i[1] = i[1].replace("blest","blessed")
        print(i[1])
```

```python
#The same but no conditionals

import csv

header = ['First_lines', 'Pron_rhymes', 'DM_narrow', 'DM_broad', 'Same_end']
with open('test.csv','a') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(header)


counter=0

while counter < len(end_words):

    line_in_question = []
    rhyme_end_words_list = []
    sort_of_list = []
    might_not_list = []
    end_same_list = []
        
    line_in_question.append(lines[counter])
    rhymes = []
    rhymes.append(pronouncing.rhymes(end_words[counter][1]))
    rhymes.append(end_words[counter][0])

    for i in rhymes[0]:
        for n in range(len(end_words)):
            if (i == end_words[n][1]):
                rhyme_end_words_list.append(end_words[n])
    #print(f'The lines in rhyme end words list that rhyme with line {rhymes[1]}, ({lines[counter]}) are {rhyme_end_words_list}')

    new_rhymes = []
    true_dm_rhymes = []
    test_dm = []
    pattern = re.compile("^.+(IY1|IY0)\s(S|Z)\s\']$")
    dm_rhymes = requests.get(f'https://api.datamuse.com/words?sl={end_words[counter][1]}&md=r')
    new_rhymes = dm_rhymes.json()
    new_rhymes.pop(0)

    for i in range(len(new_rhymes)):
        test_dm.append(new_rhymes[i].get('word'))    
    
    for i in range(len(new_rhymes)):
        if pattern.match(str(new_rhymes[i].get('tags'))):
            true_dm_rhymes.append(new_rhymes[i].get('word'))    
    
    for i in true_dm_rhymes:
        for n in range(len(end_words)):
            if (i == end_words[n][-1]):
                sort_of_list.append(end_words[n])
    #print(f'The lines in rhyme end words list that SORT OF rhyme with line {rhymes[1]}, ({lines[counter]}) are {sort_of_list}')
                
    for i in range(len(new_rhymes)):
        for n in range(len(end_words)):
            if (new_rhymes[i].get('word') == end_words[n][-1]):
                might_not_list.append(end_words[n])
    #print(f'The lines in rhyme end words list that MIGHT BUT MIGHT NOT rhyme with line {rhymes[1]}, ({lines[counter]}) are {might_not_list}')
    
    last_four = end_words[counter][1][-4:]
    last_three = end_words[counter][1][-3:]
    for n in range(len(end_words)):
        if ((last_four == end_words[n][1][-4:]) or (last_three == end_words[n][1][-3:])) and (end_words[counter][0] is not end_words[n][0]):
            end_same_list.append(end_words[n])
    #print(f'The lines in rhyme end words list that END IN THE SAME FOUR LETTERS as line {rhymes[1]}, ({lines[counter]}) are {end_same_list}')
                
    with open('test.csv','a') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerow([line_in_question, rhyme_end_words_list, sort_of_list, might_not_list, end_same_list])

    counter+=1
```
